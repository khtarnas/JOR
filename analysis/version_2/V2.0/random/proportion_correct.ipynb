```python
def get_proportion_correct(df):
    # Find total rows that have to do with testing
    df = df[df['correct_side'].notnull()]
    total = len(df.index)
    #print(total)
    
    # Find total rows where the answer was correct in testing
    correct = df[df['correct_side'] == df['chosen_side']]
    corr = len(correct.index)
    #print(corr)
    
    # Find comparison types that are in the correctly answered section:
    # print('\nThe correct comparisons are as follows:')
    # print(list(correct['comparison_type']))
    # print('\n')
    print('Proportion Correct is:')
    print(corr/total)
    # print('\n')
```

```python
'''
Function that checks whether or not a list exists in a list of lists
'''
def list_present(toplist, sublist):
    for i in toplist:
        if sublist == i:
            return True
    return False
```

```python
#set(tuple(x) for x in df['comparison_type'])
#[list(item) for item in ]#
comparisons = []
for x in list(df['comparison_type']):
    #print(type(x))
    if type(x) == list:
        if not list_present(comparisons, x):
            comparisons.append(x)

print(comparisons)
len(comparisons)
```

```python
def paired_sort(sort, follow):
        
        # Exception
        if len(sort) != len(follow):
            raise ValueError("Lists must be the same length...")
        
        comps = []
        completed = []
        for index, i in enumerate(sort):
            put = False
            
            for jndex, j in enumerate(completed):
                if i < j[0]:
                    completed.insert(jndex, [i])
                    
                    comps.insert(jndex, [follow[index]])
                    
                    put = True
                    break
                
                if i == j[0]:
                    temp = completed.pop(jndex)
                    temp.append(i)
                    completed.insert(jndex, temp)
                    
                    temp = comps.pop(jndex)
                    temp.append(follow[index])
                    comps.insert(jndex, temp)
                    
                    put = True
                    break
                    
            if not put:
                completed.append([i])
                comps.append([follow[index]])
        
        return (completed, comps)
```

```python
amount_corr = []
for comp in comparisons:
    ordered = df[df['left_comp_type'] == comp[0]]
    ordered = ordered[ordered['right_comp_type'] == comp[1]]
    inverse = df[df['left_comp_type'] == comp[1]]
    inverse = inverse[inverse['right_comp_type'] == comp[0]]
    temp = pd.concat([ordered, inverse])
    corr = temp[temp['correct_side'] == temp['chosen_side']]
    amount_corr.append(len(corr)/len(temp))
#x = dict(zip(comparisons, amount_corr))
#x

amount_corr, comparisons = paired_sort(amount_corr, comparisons)
amount_corr.reverse()
comparisons.reverse()

for i in range(len(amount_corr)):
    print(str(comparisons[i]) + ': ' + str(amount_corr[i]))
    print('\n')
#amount_corr
```
