```python
'''
Function that checks whether or not a list exists in a list of lists
'''
def list_present(toplist, sublist):
    for i in toplist:
        if sublist == i:
            return True
    return False
```

```python
def mergesort(l):
    
    length = len(l)
    if length <= 1:
        return l
    
    else:
        first = mergesort(l[:length//2])
        second = mergesort(l[length//2:])
        i = j = 0
        s = [] #new sorted array
        
        # Merge by comparing the smallest (unused) elements in each list
        while i < len(first) and j < len(second):
            if first[i] <= second[j]:
                s.append(first[i])
                i += 1
            else:
                s.append(second[j])
                j += 1
        
        # Add in all remaining elements from 'first'
        while i < len(first):
            s.append(first[i])
            i += 1
        
        # Add in all remaining elements from 'second'
        while j < len(second):
            s.append(second[j])
            j += 1

        return s
```

```python
comparisons = []
for x in list(df['comparison_type']):
    if type(x) == list:
        if not list_present(comparisons, x):
            comparisons.append(x)
print(comparisons)
```

```python
#Sort by the separation between comparisons, if separation is the same, sort by most recent
def separation_sort(comps):
    sep = []
    for i in comps:
        #print(i)
        if len(sep) == 0:
            sep.append(i)
        else:
            temp = sep.copy()
            for idx, j in enumerate(sep):
                #print('before')
                #print(temp)
                if (i[1] - i[0]) < (j[1] - j[0]): #if the new separation is less than the one in this position in sep
                    temp.insert(idx, i)
                    #print('less than')
                    #print(temp)
                    break
                elif (i[1] - i[0]) == (j[1] - j[0]):#else if same, check first val
                    if (i[0] < j[0]):
                        temp.insert(idx, i)
                        #print('equal then less than')
                        #print(temp)
                        break
                elif (idx == len(sep) - 1):
                    temp.append(i)
                    break
            sep = temp.copy()
            #print(sep)
    return sep
```

```python
comparisons = separation_sort(comparisons)
```

```python
distances = []
for comp in comparisons:
    distances.append(comp[0])
    distances.append(comp[1])
distances = list(set(distances))
distances = mergesort(distances)
distances
```

```python
def average_for_comp(df, comp):
    ordered = df[df['left_comp_type'] == comp[0]]
    ordered = ordered[ordered['right_comp_type'] == comp[1]]
    inverse = df[df['left_comp_type'] == comp[1]]
    inverse = inverse[inverse['right_comp_type'] == comp[0]]
    temp = pd.concat([ordered, inverse])
    corr = temp[temp['correct_side'] == temp['chosen_side']]
    rt = temp['rt'].mean()
    return rt
```

```python
def graph_for_distance(distance, comparisons, distances, df):
    
    # for each comparison for this distance (i.e. 4 and 16) add the average to the y array in order (FOR COMP = DISTANCE, set to 0.5)
    y1 = []
    y2 = []
    x1 = []
    x2 = []
    
    passed = False
    for dist in distances:
        if dist == distance:
            passed = True
        elif not passed:
            y1.append(average_for_comp(df, [distance, dist]))
            x1.append(dist)
        elif passed:
            y2.append(average_for_comp(df, [distance, dist]))
            x2.append(dist)
            
    # Set up graph
    # TEMPORARY SET UP
    fig, ax = plt.subplots()
    plt.title('Response Time for Distance of ' + str(distance) + ' Against Other Distances', pad=20)
    plt.ylabel('Response Time')
    plt.xlabel('Distance Compared To')
    plt.xticks(distances)
    plt.xlim(3, 41)
    #plt.ylim(2750, 9000)
    #plt.yticks(np.linspace(0,1,11))
    ax.yaxis.set_ticks_position('both')
    ax.plot(x1, y1, clip_on = False, color = 'black', marker = 'o')
    ax.plot(x2, y2, clip_on = False, color = 'black', marker = 'o', label=distance)
    
    idx = distances.index(distance)
    
    if distance != 4 and distance != 6:
        separation = 2
        if distance >= 14:
            separation = 4
            if distance > 22:
                separation = 6
        end_pos = distances[idx - 1]
        p1 = np.polyfit(x1, y1, deg=1)
        ax.plot(np.arange(4, distance, separation), np.polyval(p=p1, x=np.arange(4, distance, separation)), color='cornflowerblue')
    
    if distance != 40 and distance != 34:
        start_pos = distances[idx + 1]
        p2 = np.polyfit(x2, y2, deg=1)
        ax.plot(np.arange(start_pos, 42, 2), np.polyval(p=p2, x=np.arange(start_pos, 42, 2)), color='cornflowerblue')
    
    plt.savefig('rt_for_' + str(distance) +'.pdf', format='pdf', dpi=1200, bbox_inches='tight')
    plt.show()

for i in distances:
    graph_for_distance(i, comparisons, distances, df)
```
